## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

## JSON exercise

Using data in file 'data/world_bank_projects.json' and the techniques demonstrated above,
1. Find the 10 countries with most projects
2. Find the top 10 major project themes (using column 'mjtheme_namecode')
3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
import pandas as pd
import json
from pandas.io.json import json_normalize

#First import the file
port = "C:/Users/joshu/Downloads/data_wrangling_json/world_bank_projects.json"
json_df = pd.read_json(port)
```

## 1. Find the 10 countries with most projects

```python
#Check to see what the columns are
json_df.columns
```

```python
#Using 'countryname', use the groupby method to count the number of row multiples 
country_multiples = json_df.groupby('countryname')['countryname'].count()
top10_countries = country_multiples.sort_values(ascending=False)
print(top10_countries[0:10])
```

## 2. Find the top 10 major project themes (using column 'mjtheme_namecode')

```python
# Normalize the file to get the project theme names
# Need to first read the file as a string 'code'
with open(port) as port2:
    code = json.loads(port2.read())
json_themes = json_normalize(code, 'mjtheme_namecode')
json_themes.head()
```

```python
# Check to see if there are any null values. 
json_themes.info()
```

```python
# Verify if there are any missing entries for code
json_themes[json_themes['code']=='']
```

```python
# Verify if there are any missing entries for name
json_themes[json_themes['name']=='']
```

```python
# Since there are no missing entries for 'code', use 'code' to count the number of themes 
themes_count = json_themes.groupby('code').count()
themes_count.columns=['Group Count']
themes_count = themes_count.sort_values(ascending=False, by='Group Count')
themes_count[0:10]
```

## 3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
# One way to fill in the missing names is to first sort by code and then to fill in the name by using fill forward
new_df = json_themes.sort_values(ascending=False, by='code')
new_df[new_df['name']==''].replace('','None')
new_df.fillna(method='ffill')
new_df
```
